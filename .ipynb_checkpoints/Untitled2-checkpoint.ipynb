```python
import os
path = "./data"

articles = []
for root,d_names,f_names in os.walk(path):
    print(root, d_names, f_names)
    for name in f_names:
        article = open(root + '/' + name, 'r')
        article_str = article.read()
        articles.append(article_str)
```

```python
articles[100008]
```

```python
from bs4 import BeautifulSoup
```

```python
import re
```

```python
cleaned_articles = []
for article in articles:
    article = article.replace('\n',' ')
    article = ' '.join(article.split())
    #article = article.lower()
    article = re.sub("<.*>", "", article)
    article = re.sub("\[[a-zA-Z0-9]*\]", "", article)
    article = re.sub("Luokat: .*", "", article)
    article = re.sub("Luokka: .*", "", article)
    article = re.sub("http[^ ]*", "", article)
    article = re.sub("[|:/]", " ", article)
    article = re.sub("Tm .* artikkeli on tynk. .* laajentamalla artikkelia.", "", article)
    article = re.sub("Tm artikkeli .* tarvitsee laajennusta.", "", article)
    article = re.sub("Tm on tsmennyssivu[^\.]*.", "", article)
    article = re.sub("Katso mys luetteloa tsmennyssivuista.", "", article)
    article = re.sub("Aiheesta muualla", "", article)
    article = ' '.join(article.split())
    
    cleaned_articles.append(article)
    
```

```python
from fastai.text import *
```

```python
df = DataFrame(data=cleaned_articles)
```

```python
df['label'] = 0
```

```python
df.columns = ['text', 'label']
```

```python
df = df[['label', 'text']]
```

```python
df
```

```python
train_df = df[0:80000]
```

```python
val_df = df[80000:]
```

```python
val_df
```

```python
data_lm = TextLMDataBunch.from_df('./data/lm', train_df, val_df)
```

```python
data_lm.save('data_lm_export.pkl')
```

```python
learn = language_model_learner(data_lm, AWD_LSTM, drop_mult=0.5)
```

```python
learn.fit_one_cycle(3, 1e-2)
```

```python
learn.unfreeze()
learn.fit_one_cycle(9, 1e-3)
learn.save_encoder('ft_enc')
```

```python
learn.predict("Tuomas Jokioja", n_words=50)
```
